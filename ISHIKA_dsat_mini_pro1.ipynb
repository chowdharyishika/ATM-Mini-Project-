# 1. Documentation

Automated Teller Machines(ATM)  allows customers to complete basic transactions without the aid of a branch representative or teller. This project describes the working of ATM. It allows the users to perform certain ... such as check saving balance, deposit funds from the user, withdraw funds from the savings balance account and transfer funds from checking balance to savings balance.

A class ATM is created. It has the following methods:

1. __init__(self,acc_bal,check_bal,name,pin): It initializes the values of:
   self.acc_bal: saving account balance
   self.check_bal: checking balance
   self.name: Name of the account holder
   self.pin: Pin of the ATM card

2. check_saving_balance(self): It returns the current saving balance

3. deposit_funds(self): It asks the user to enter the amount to be deposited (funds) and adds it to the saving balance. It     also prints the updated saving balance.

4. withdraw_funds(self): It asks the user to enter the amount to be withdrawn (funds) and checks if the account holds sufficient balance for withdrawal. If the amount is insufficient, it gives warning "Insufficient balance" else subtracts the amount from saving balance. It also prints the updated saving balance.

5. transfer_funds(self): It asks the user to enter the amount to be withdrawn (amount) from the checking balance and checks if the checking balance is sufficient for withdrawal. If the amount is insufficient, it gives warning "Insufficient balance" else subtracts the amount from checking balance and adds it into saving balance. It also prints the updated saving balance and updated checking balance.

6. main(self): It is the main function. It asks the user for pin , two times for double authentication. If the pin entered by user is incorrect, it gives the message "Incorrect pin" else, it asks the user for choice 'ch' (1 to 4) and calls the particular methods according to the choice:
    For ch=1, check_saving_balance(self) is called.
    For ch=2, deposit_funds(self) is called.
    For ch=3, withdraw_funds(self) is called.
    For ch=4, transfer_funds(self) is called.
    If user enters choice other than 1 to 4, warning is given "Invalid option".

WORKING OF THE PROJECT:
User is asked for the following choice(c):
 c=1 to Register
 c=2 to Login
 c=3 to Exit

1. Register : It is called to generate a pin. In this user is asked for the following details:
   name:Name    
   pin: Pin
   acc_bal: Initial savings balance
   check_bal: Initial checking balance
   acc_no: Account number
   It then adds the details in the dictionary of existing user(d) with account number as the key

2. Login : It is called to perform the above mentioned actions by the existing users. It asks for the account number(apinn). If     the account number is not in the dictionary, it gives the warning "Invalid account number" else creates and object of class     ATM and calls the main function.

3. Exit : It is used to terminate the program.

4. If user input choice other than 1 to 3, it gives the warning message "Invalid choice".



# 2. Source Code

```python
class ATM:
    def __init__(self,acc_bal,check_bal,name,pin):  #initializes the values of attributes
        self.acc_bal=acc_bal
        self.check_bal=check_bal
        self.name=name
        self.pin=pin
        
    def check_saving_balance(self):           #to Check the saving balance
        return self.acc_bal         
    
    def deposit_funds(self):                 #to Deposit funds from the user
        funds=int(input("Enter the amount"))
        self.acc_bal+=funds                      
        print("Saving balance is:",self.acc_bal)
        
    def withdraw_funds(self):             #to Withdraw funds from the savings balance account
        funds=int(input("Enter the amount"))
        if funds>self.acc_bal:    #alerts the user that he/she is trying to withdraw more than the amount present in the account                   
            print("Insufficient balance")
        else:
            self.acc_bal-=funds
            print("Savings balance is:",self.acc_bal)
            
    def transfer_funds(self):           #to Transfer funds from checking balance to savings balance
        amount=int(input("Enter the amount"))
        if self.check_bal<amount: #alerts the user that he/she is trying to withdraw more than the amount present in the account
            print("Insufficient balance")
        else:
            self.acc_bal+=amount
            self.check_bal-=amount
            print("Savings balance is:",self.acc_bal)
            print("Checking balance is:",self.check_bal)
        
    def main(self):
        f=1
        while f==1:  
            
            print("To exit press 0 else press 1")    #Asks if user want to continue or exit
            
            z=int(input())
            if z==0:
                f=0
                break
    
            p=input("Enter the pin")
            pin=input("Enter the pin again for double authentication")
            
            if pin==self.pin and p==self.pin:    #checking if pin is correct or not
                print("Press 1 to Check the saving balance",end="\n")
                print("Press 2 to Deposit funds from the user",end="\n")
                print("Press 3 to Withdraw funds from the savings balance account",end="\n")
                print("Press 4 to Transfer funds from checking balance to savings balance",end="\n")
                
                ch=int(input("Enter choice"))  #input the existing users choice for further actions

                if ch==1:
                    a=atm.check_saving_balance()
                    print("Saving balance is :",a)
                    
                elif ch==2:
                    atm.deposit_funds() 
                    
                elif ch==3:
                    atm.withdraw_funds()
                    
                elif ch==4:
                    
                    atm.transfer_funds()
                else:
                    print("Invalid option")   #warning message
                    f=0
                    break
            else:
                print("Incorrect Pin")  #warning message
                f=0
                break
                
k=1
d={} #Dictionary of existing user
l=0
print("WELCOME")
while k==1:
    
    print("Press 1 to Register",end="\n")
    print("Press 2 to Login",end="\n")
    print("Press 3 to Exit",end="\n")
    
    c=int(input("Enter choice"))
    
    if c==1:         #registers the new user and input all his/her required information
        name=input("Input name")    
        pin=input("Generate four digit pin")
        acc_bal=int(input("Input initial savings balance"))
        check_bal=int(input("Input initial checking balance"))
        acc_no=int(input("Input account number"))
        
        d[acc_no]=[name,pin,acc_bal,check_bal]
        
    elif c==2:  #logins the existing user for further actions
        apinn=int(input("Enter account number"))
        
        if apinn in d.keys():
            atm=ATM(d[apinn][2],d[apinn][3],d[apinn][0],d[apinn][1])
            atm.main()
            
        else:
            print("Invalid account number")   #warning message
            
    elif c==3:
        k=0
        break
        
    else:
        print("Invalid choice")   #warning message
```
